```python
## 교수님
class Student(object):
    
    # class variable
    
    scholarship = 0.5
    stuCount = 0
    tuitionFee = 3000000
    
    def __init__(self,id,name,email,score):
        self.id=id
        self.name=name
        self.email=email
        self.score=score
        Student.stuCount+=1 # 학생 수 count
        
    def displayInfo(self):
        return 'Student Information: {}, {}, {}, {}'.format(self.id, self.name, self.email, self.score)
    
    def __str__(self):
        return 'Student: {} {}'.format(self.id, self.email) # print했을 때 나왔으면 하는 form: __str__
    
    def calcuAvg(self):
        avg = sum(self.score)/len(self.score)
        print('name: {}, 평균: {}'.format(self.name,avg))
    
    def chkScholarship(self):
        if (sum(self.score)/len(self.score))>=90:
            print('{} is a scholarship student'.format(self.name))
        else:
            print('{} is not a scholarship student'.format(self.name))
            
    def calcuTuition(self):
        if (sum(self.score)/len(self.score))>=90:
            return 'name: {}, fee: {}'.format(self.name,float(Student.scholarship)*int(Student.tuitionFee))
        else:
            return 'name: {}, fee: {}'.format(self.name,int(Student.tuitionFee)) # 굳이 tuitionFee=1로 설정할 필요 없다.
        
    def stuCounts(self):
        print('전체 학생수: {}'.format(Student.stuCount))
        
stu1 = Student(1,'이순신','stu1@hanyang.ac.kr',[90,80,100])
stu2 = Student(2,'강감찬','stu2@seoul.com',[70,80,90])

print(stu1)
print(stu2)

print(stu1.displayInfo())
print(stu2.displayInfo())

stu1.chkScholarship()
stu2.chkScholarship()

stu1.calcuAvg()
stu2.calcuAvg()

print(stu1.calcuTuition())
print(stu2.calcuTuition())

stu2.stuCounts()
```
